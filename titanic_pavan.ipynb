# Data Cleaning

**Check for duplicates and handle them**

**check for missing data and handle them**

**check for invalid dtypes and handle them**

```python
df.duplicated().any() # no duplicates records
```

```python
df.info()
```

```python
age_mean=df['age'].mean()
fare_mean=df['fare'].mean()
```

```python
emb_mode=df['embarked'].mode()[0]
```

```python
df['age']=df['age'].fillna(age_mean)
df['fare']=df['fare'].fillna(fare_mean)
df['embarked']=df['embarked'].fillna(emb_mode)
```

```python
df=df.dropna(axis=1)
```

```python
df.info() # no missing data and no invalid data
```
