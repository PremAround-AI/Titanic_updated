#Apply Function

```python
def comp_type(row):
    f=row['fam_cnt']
    p=row['no_of_psgs']

    if p==1:
        return 'solo'
    elif f==0 and p>1:
        return 'friends'
    else:
        return 'family'
```

```python
df['companion type']=df.apply(comp_type,axis=1)  #here apply func is giving args as row,
                                                 #internally by default 
```

```python
df['companion type']
```

#Cut Method likewise Apply method

```python
pd.cut(df['age'],bins=[0,2,16,25,59,100],labels=['Infant','child','teen','adult','senior'])
```

```python
df['age_category']=pd.cut(df['age'],bins=[0,2,16,25,59,100],labels=['Infant','child','teen','adult','senior'])
df.head(3)
```

```python
df.courtsey_title.unique()
```

```python
royal=['the Countess','Lady','Sir','Don','Dona','Jonkheer']
common=['Miss','Mr', 'Mrs','Master','Ms', 'Mme','Mlle']
military=['Col','Major','Capt']
others=['Rev','Dr']

def social_status(row):
    c=row['courtsey_title']
    if c in royal:
        return 'royal'
    elif c in common:
        return 'common'
    elif c in military:
        return 'military'
    else:
        return 'others'


df['social status']=df.apply(social_status,axis=1)
df.head()
```
